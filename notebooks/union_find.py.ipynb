```python
import os
import zipfile

import pandas as pd

from text_dedup.utils.union_find import UnionFind as PythonUnionFind
from text_dedup.utils.union_find_rust import UnionFind as RustUnionFind
```

```python
if not os.path.exists("twitch_gamers.zip"):
    !wget https://snap.stanford.edu/data/twitch_gamers.zip
    with zipfile.ZipFile("twitch_gamers.zip", "r") as zip_ref:
        zip_ref.extractall("twitch_gamers")
```

```python
df = pd.read_csv("twitch_gamers/large_twitch_edges.csv")
df.head()

edges = df.values.tolist()
nodes = set(df.values.reshape(-1).tolist())
```

```python
MAX_SIZE = len(nodes) + 1
print(f"Found {len(nodes) - 1} nodes")
```

```python
%timeit

uf_python = PythonUnionFind()
for x, y in edges:
    uf_python.union(x, y)

for node in nodes:
    uf_python.find(node)
```

```python
%timeit

uf_rust = RustUnionFind(max_size=MAX_SIZE)

for x, y in edges:
    uf_rust.union(x, y)

for node in nodes:
    uf_rust.find(node)
```

```python
for node in nodes:
    if not uf_rust.same(uf_rust.find(node), uf_python.find(node)):
        print(f"Mismatch node: {node}")
```
